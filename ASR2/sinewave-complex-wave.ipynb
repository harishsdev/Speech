```python
import matplotlib.pylab as plt
```

```python
import numpy as np
```

```python
A=0.8
```

```python
f0=1000
```

```python
phi=np.pi/2
```

```python
fs=44100
```

```python
t=np.arange(-0.02,0.02,1.0/fs)
```

```python
t
```

```python
x=A*np.cos(2*np.pi*f0*t+phi)
```

```python
x
```

```python
plt.plot(t,x)
```

```python
plt.axis([-0.002,0.002,-.8,.8])
plt.xlabel('time')
plt.ylabel('amplitude')
plt.plot(t,x)
```

```python
#N=500
#k=3
N=32
k=5
n=np.arange(-N/2,N/2)
```

```python
s=np.exp(1j*2*np.pi*k*n/N)
```

```python
plt.plot(n,np.real(s))
#plt.axis(-N/2,N/2-1,-1,1)
plt.xlabel('n')
plt.ylabel('amplitude')
```

```python
plt.plot(n,np.imag(s))
#plt.axis(-N/2,N/2-1,-1,1)
plt.xlabel('n')
plt.ylabel('amplitude')
```
